# questão 1

```python
custo = float(input("Custo de fábrica: R$ "))

# Calcular impostos (45%) e distribuição (28%)
impostos = custo * 0.45
custo_com_impostos = custo + impostos

distribuidor = custo_com_impostos * 0.28
custo_final = custo_com_impostos + distribuidor

# Mostrar resultado final
print(f"Custo ao consumidor: R$ {custo_final:.2f}")
```

# questão 2

```python
# Solicita as horas e minutos separadamente
horas = int(input("Digite as horas: "))
minutos = int(input("Digite os minutos: "))

# Converte minutos em fração de hora
minutos_decimal = minutos / 60

# Soma as horas com a fração dos minutos
horas_decimais = horas + minutos_decimal

# Converte tudo para segundos (1 hora = 3600 segundos)
segundos = horas_decimais * 3600

# Exibe os resultados
print(f"\nHoras decimais: {horas_decimais:}")
print(f"Total em segundos: {segundos:}")
```

# questão 3

```python
# Entrada: tempo trabalhado
horas = int(input("Horas extras trabalhadas: "))
minutos = int(input("Minutos extras trabalhados: "))

# Converte minutos em fração de hora
minutos_decimal = minutos / 60
horas_totais = horas + minutos_decimal

# Entrada: valor ganho por hora normal
valor_por_hora = float(input("Valor ganho por hora (R$): "))

# Hora extra tem acréscimo de 70%
valor_hora_extra = valor_por_hora * 1.70

# Cálculo do valor total das horas extras
valor_total = horas_totais * valor_hora_extra

# Saída
print(f"\nHoras extras em formato decimal: {horas_totais:.2f}")
print(f"Valor total a receber pelas horas extras: R$ {valor_total:.2f}")
```

# quetão 4

```python
capital = float(input("Digite o valor do investimento (R$): "))
taxa_juros = float(input("Digite a taxa de juros anual (%): "))
anos = int(input("Digite o período do investimento (em anos): "))

# Converte a taxa de juros de porcentagem para decimal
i = taxa_juros / 100

# Aplica a fórmula dos juros compostos
montante = capital * (1 + i) ** anos

# Saída do valor final
print(f"\nValor final após {anos} anos: R$ {montante:.2f}")
```

# questão 5

```python
# Entrada de dados
valor_conta = float(input("Digite o valor total da conta (R$): "))
gorjeta = float(input("Digite o valor da gorjeta (R$): "))
pessoas = int(input("Quantas pessoas vão dividir a conta? "))

# Calcula taxa de serviço (10% da conta)
taxa_servico = valor_conta * 0.10

# Soma tudo: conta + taxa de serviço + gorjeta
valor_total = valor_conta + taxa_servico + gorjeta

# Calcula quanto cada pessoa deve pagar
valor_por_pessoa = valor_total / pessoas

# Exibe os resultados
print(f"\nValor da conta: R$ {valor_conta:.2f}")
print(f"Taxa de serviço (10%): R$ {taxa_servico:.2f}")
print(f"Gorjeta: R$ {gorjeta:}")
print(f"Valor total com taxas: R$ {valor_total:.2f}")
print(f"Cada pessoa deve pagar: R$ {valor_por_pessoa:.2f}")
```
